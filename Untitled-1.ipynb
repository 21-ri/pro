Importing Datasets

```python
import pandas as pd
import matplotlib.pyplot as plt
from pandas.plotting import lag_plot
from statsmodels.tsa.arima_model import ARIMA
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
from statsmodels.tsa.stattools import adfuller
```

```python
df = pd.read_csv('/Users/parthshah/Downloads/sm/Processed_NYSE.csv' ,index_col='Date', parse_dates=['Date'])

df.head()
```

```python
df1 = pd.read_csv('/Users/parthshah/Downloads/sm/Processed_RUSSELL.csv' ,index_col='Date', parse_dates=['Date'])
df1.tail()
```

Cleaning and preparing time series data

```python
df1 = df1.iloc[1:]
df1 = df1.fillna(method='ffill')
df1.head()
```

Create a date range

```python
start_date = '2009-12-31'
end_date = '2010-01-31'
date_range = pd.date_range(start=start_date, end=end_date)
```

Visualizing the datasets

Plot lag plots and Timestamps

```python
plt.figure(figsize=(12, 6))
lag_plot(df['Close'])
plt.title('Lag Plot for NYSE Dataset')
plt.show()
```

```python
plt.xlabel('Timestamp')
plt.ylabel('Lagged Close Value')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
lag_plot(df1['Close'])
plt.title('Lag Plot for RUSSELL Dataset')
plt.show()
```

```python
plt.xlabel('Timestamp')
plt.ylabel('Lagged Close Value')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
plot_acf(df['Close'], lags=20)
plt.title('ACF for NYSE Dataset')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
plot_pacf(df1['Close'], lags=20)
plt.title('PACF for RUSSELL Dataset')
plt.show()
```

Perform Augmented Dickey-Fuller test

```python
result = adfuller(df['Close'])
print("ADF Statistic:", result[0])
print("p-value:", result[1])
```

```python
result1 = adfuller(df1['Close'])
print("ADF Statistic:", result1[0])
print("p-value:", result1[1])
```
