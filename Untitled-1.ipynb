## MA models

```python
np.random.seed(42)
n = 200
time = np.arange(n)
simulated_data1 = 10 + 0.5 * time + np.random.normal(0, 1, n)
simulated_data2 = 15 - 0.3 * time + np.random.normal(0, 1, n)

```

```python
simulated_df1 = pd.DataFrame({'Date': pd.date_range(start='2025-01-01', periods=n), 'Value': simulated_data1})
simulated_df2 = pd.DataFrame({'Date': pd.date_range(start='2025-01-01', periods=n), 'Value': simulated_data2})
```

```python
train_size = int(len(simulated_df1) * 0.8)
train_df1, test_df1 = simulated_df1.iloc[:train_size], simulated_df1.iloc[train_size:]
train_df2, test_df2 = simulated_df2.iloc[:train_size], simulated_df2.iloc[train_size:]
```

```python
train_values1 = train_df1['Value']
test_values1 = test_df1['Value']
train_values2 = train_df2['Value']
test_values2 = test_df2['Value']
```

```python
model1 = auto_arima(train_values1, d=0, seasonal=False, trace=True)
model2 = auto_arima(train_values2, d=0, seasonal=False, trace=True)
```

```python
forecast_steps = len(test_values1)
forecast1 = model1.predict(n_periods=forecast_steps)
forecast2 = model2.predict(n_periods=forecast_steps)
```

```python
from math import sqrt
```

```python
rmse1 = sqrt(mean_squared_error(test_values1, forecast1))
rmse2 = sqrt(mean_squared_error(test_values2, forecast2))
print("RMSE for Dataset 1:", rmse1)
print("RMSE for Dataset 2:", rmse2)
```

```python
simulated_df1 = pd.DataFrame({'Date': pd.date_range(start='2025-01-01', periods=n), 'Value': simulated_data1})
simulated_df2 = pd.DataFrame({'Date': pd.date_range(start='2025-01-01', periods=n), 'Value': simulated_data2})
```

```python
train_size = int(len(simulated_df1) * 0.8)
train_df1, test_df1 = simulated_df1.iloc[:train_size], simulated_df1.iloc[train_size:]
train_df2, test_df2 = simulated_df2.iloc[:train_size], simulated_df2.iloc[train_size:]
```

```python
train_values1 = train_df1['Value']
test_values1 = test_df1['Value']
train_values2 = train_df2['Value']
test_values2 = test_df2['Value']
```

```python
model1 = auto_arima(train_values1, trace=True)
model2 = auto_arima(train_values2, trace=True)
```

```python
forecast_steps = len(test_values1)
forecast1 = model1.predict(n_periods=forecast_steps)
forecast2 = model2.predict(n_periods=forecast_steps)
```

```python
rmse1 = sqrt(mean_squared_error(test_values1, forecast1))
rmse2 = sqrt(mean_squared_error(test_values2, forecast2))
print("RMSE for Dataset 1:", rmse1)
print("RMSE for Dataset 2:", rmse2)
```

```python
plt.figure(figsize=(12, 6))
plt.plot(test_df1['Date'], test_values1, label='Actual  MacDonalds data')
plt.plot(test_df1['Date'], forecast1, color='red', label='Forecasted MacDonalds data')
plt.plot(test_df2['Date'], test_values2, label='Actual Dominos data')
plt.plot(test_df2['Date'], forecast2, color='green', label='Forecasted Dominos data')
plt.legend()
plt.title('ARMA Model Forecasting for Two Datasets')
plt.show()
```

## LSTM Model

```python
np.random.seed(42)
n = 200
time = np.arange(n)
simulated_data1 = 10 + 0.5 * time + np.random.normal(0, 1, n)
simulated_data2 = 15 - 0.3 * time + np.random.normal(0, 1, n)
```

```python
simulated_df1 = pd.DataFrame({'Date': pd.date_range(start='2025-01-01', periods=n), 'Value': simulated_data1})
simulated_df2 = pd.DataFrame({'Date': pd.date_range(start='2025-01-01', periods=n), 'Value': simulated_data2})
```

```python
train_size = int(len(simulated_df1) * 0.8)
train_df1, test_df1 = simulated_df1.iloc[:train_size], simulated_df1.iloc[train_size:]
train_df2, test_df2 = simulated_df2.iloc[:train_size], simulated_df2.iloc[train_size:]
```

```python
train_values1 = train_df1['Value'].values.reshape(-1, 1)
test_values1 = test_df1['Value'].values.reshape(-1, 1)
train_values2 = train_df2['Value'].values.reshape(-1, 1)
test_values2 = test_df2['Value'].values.reshape(-1, 1)
```

```python
scaler = MinMaxScaler()
train_values1 = scaler.fit_transform(train_df1['Value'].values.reshape(-1, 1))
test_values1 = scaler.transform(test_df1['Value'].values.reshape(-1, 1))
train_values2 = scaler.fit_transform(train_df2['Value'].values.reshape(-1, 1))
test_values2 = scaler.transform(test_df2['Value'].values.reshape(-1, 1))
```

```python
def create_sequences(data, sequence_length):
    sequences = []
    for i in range(len(data) - sequence_length + 1):
        sequences.append(data[i:i + sequence_length])
    return np.array(sequences)
```

```python
sequence_length = 10
n_features = 1 
```

```python
train_sequences1 = create_sequences(train_values1, sequence_length)
test_sequences1 = create_sequences(test_values1, sequence_length)
train_sequences2 = create_sequences(train_values2, sequence_length)
test_sequences2 = create_sequences(test_values2, sequence_length)
```

```python

model = Sequential()
model.add(LSTM(50, activation='relu', input_shape=(sequence_length, n_features)))
model.add(Dense(1))
model.compile(optimizer='adam', loss='mean_squared_error')
```

```python
model.fit(train_sequences1, train_sequences1[:, -1], epochs=50, batch_size=32)
```

```python
test_predictions1 = model.predict(test_sequences1)
test_predictions1 = scaler.inverse_transform(test_predictions1)  
```

```python
test_predictions2 = model.predict(test_sequences2)
test_predictions2 = scaler.inverse_transform(test_predictions2)  
```

```python
rmse1 = sqrt(mean_squared_error(test_values1[sequence_length - 1:], test_predictions1))
print("RMSE for Dataset 1:", rmse1)
```

```python
rmse2 = sqrt(mean_squared_error(test_values2[sequence_length - 1:], test_predictions2))
print("RMSE for Dataset 2:", rmse2)
```

```python
absolute_percentage_errors = np.abs((test_values1[sequence_length - 1:] - test_predictions1) / test_values1[sequence_length - 1:]) * 100
```

```python
absolute_percentage_errors1 = np.abs((test_values2[sequence_length - 1:] - test_predictions2) / test_values2[sequence_length - 1:]) * 100
```

```python
plt.figure(figsize=(12, 6))

plt.plot(test_df1['Date'][sequence_length - 1:], test_values1[sequence_length - 1:], label='Actual MaccDonalds data')
plt.plot(test_df1['Date'][sequence_length - 1:], test_predictions1, color='red', label='Forecasted MacDOnalds data')
plt.plot(test_df2['Date'][sequence_length - 1:], test_values2[sequence_length - 1:], label='Actual Dominos data')
plt.plot(test_df2['Date'][sequence_length - 1:], test_predictions2, color='green', label='Forecasted Dominos data')

plt.legend()
plt.title('LSTM Model Forecasting for MacDonalds and Dominos data')
plt.show()
```
