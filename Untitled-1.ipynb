Modelling using statstools

AR models

```python
train_size = int(len(df) * 0.8)
train_df, test_df = df.iloc[:train_size], df.iloc[train_size:]
train_df1, test_df1 = df1.iloc[:train_size], df1.iloc[train_size:]
```

```python
lag = 10
ar_model_df = AutoReg(train_df['Close'], lags=lag)
ar_result_df = ar_model_df.fit()
```

```python
ar_model_df1 = AutoReg(train_df1['Close'], lags=lag)
ar_result_df1= ar_model_df1.fit()
```

```python
predictions_df = ar_result_df.predict(start=len(train_df), end=len(train_df) + len(test_df) - 1, dynamic=False)

predictions_df1 = ar_result_df1.predict(start=len(train_df1), end=len(train_df1) + len(test_df1) - 1, dynamic=False)
```

```python
pip install scikit-learn
```

```python
from sklearn.metrics import mean_squared_error
```

```python
mse_df = mean_squared_error(test_df['Close'], predictions_df)
mse_df1 = mean_squared_error(test_df1['Close'], predictions_df1)
```

```python
plt.figure(figsize=(12, 6))
plt.plot(test_df.index, test_df['Close'], label='Actual')
plt.plot(test_df.index, predictions_df, color='yellow', label='Predicted')
plt.legend()
plt.title('AR Model Forecasting for MAcdonalds Stock Market Data')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
plt.plot(test_df1.index, test_df1['Close'], label='Actual')
plt.plot(test_df1.index, predictions_df1, color='red', label='Predicted')
plt.legend()
plt.title('AR Model Forecasting for Dominos Stock Market Data')
plt.show()
```

```python
print('Mean Squared Error for df:', mse_df)
print('Mean Squared Error for df1:', mse_df1)
```
